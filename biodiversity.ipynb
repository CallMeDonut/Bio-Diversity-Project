### Analysis

```python
##Make NaN values a column in the dataframe
species.fillna('No Intervention', inplace=True)
species.groupby('conservation_status').size()
```

```python
##Find out how many in each category are in each conservation status
conservation_category = species[species.conservation_status != 'No Intervention']\
    .groupby(['conservation_status', 'category'])['scientific_name']\
    .count()\
    .unstack()
conservation_category
```

```python
##Create stacked bar graph to show this data
label_order = ['Species of Concern', 'Threatened', 'Endangered', 'In Recovery']
ax = conservation_category.loc[label_order].plot(kind='bar', stacked=True, figsize=(8,6))
ax.set_xlabel('Conservation Status')
ax.set_ylabel('Number of Species')
```

```python
##create column showing all species that are protected (not in the "no intervention" column)
species['is_protected'] = species.conservation_status != 'No Intervention'
```

```python
##Show how many of each category is protected/not protected
category_counts = species.groupby(['category', 'is_protected'])\
    .scientific_name.nunique()\
    .reset_index()\
    .pivot(columns='is_protected',
                    index='category',
                    values='scientific_name')\
                    .reset_index()
category_counts.columns = ['category', 'not_protected', 'protected']
category_counts
```

```python
## Calculate percentage of species protected  in each category
category_counts['percent_protected'] = category_counts.protected / \
                                    (category_counts.protected + category_counts.not_protected) * 100
category_counts
```

```python
##chi2 test on mammals vs birds
from scipy.stats import chi2_contingency as chi
contingency1 = [[146, 30],
                [413, 75]]
chi(contingency1)
##No significance
```

```python
##chi2 test on mammals vs reptiles
contingency2 = [[146, 30],
                [73, 5]]
chi(contingency2)
##Statistically Significant
```

```python
##Split data into individual names
from itertools import chain
import string

def remove_punctuation(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

common_Names = species[species.category == 'Mammal']\
    .common_names\
    .apply(remove_punctuation)\
    .str.split().tolist()

common_Names[:6]
```

```python
##Get rid of duplicate words in each row
CleanRows = []

for item in common_Names:
    item = list(dict.fromkeys(item))
    CleanRows.append(item)
CleanRows[:6]
```

```python
## Put into one list
res = list(chain.from_iterable(i if isinstance(i, list) else [i] for i in CleanRows))
res[:6]
```

```python
## Count number of times each name was brought up
words_counted = []

for i in res:
    x = res.count(i)
    words_counted.append((i,x))
words_counted = pd.DataFrame(set(words_counted), columns=['Word', 'Count']).sort_values('Count', ascending=False)
words_counted.head(10)
```

```python
## Add boolean value to check if is_bat = True
species['is_bat'] = species.common_names.str.contains(r'\bBat\b', regex=True)
species.head(10)
```

```python
## All bats in the dataframe
species[species['is_bat'] == True]
```

```python
## Combine bat data with observations dataset to show observations of bats in each park
bat_observations = observations.merge(species[species.is_bat])
bat_observations
```

```python
## Show number of observations of bats in each park
bat_observations.groupby('park_name').observations.sum().reset_index()\
    .sort_values('observations', ascending=False)
```

```python
## Show number of observations of bats in each park by whether or not they're protected
obs_by_park = bat_observations.groupby(['park_name', 'is_protected']).observations.sum().reset_index()\
    .sort_values('observations', ascending=False)
obs_by_park
```

```python
## Plot above data
plt.figure(figsize=(16,4))
sns.barplot(x=obs_by_park.park_name,\
            y=obs_by_park.observations,\
            hue=obs_by_park.is_protected)
plt.xlabel('Park Name')
plt.ylabel('Number of Observations')
plt.title('Number of Bats Per Week')
plt.show()
```
